```python
trois = pd.read_sql('SELECT date, location, MAX(new_deaths_per_million ) as deces FROM  Covid19', conn)
trois
```

```python
quatre = pd.read_sql("select location, count(*), total_deaths_per_million from 'Covid19' where location == 'France' and total_deaths_per_million >= 1000", conn)
quatre
```

```python
cinq = pd.read_sql("SELECT location, total_deaths_per_million FROM Covid19 where location = 'France' OR location = 'United States' OR location = 'India';", conn)
cinq
```
